## Set Aside Test Set

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['median_house_value'], axis=1).copy()
y = df['median_house_value']
X_train, X_rem, y_train, y_rem = train_test_split(X, y, train_size=0.8, random_state=5)
```

## Split Train Set and Validation Set 

```python
X_valid, X_test, y_valid, y_test = train_test_split(X_rem, y_rem, test_size=0.5,random_state=5)

print(X_train.shape)
print(X_test.shape)
print(X_valid.shape)
```

## Define Function to Measure Quality of the Model

```python
from sklearn.metrics import mean_absolute_error
from math import sqrt
def score_data (model, X_train, X_valid, y_train, y_valid):
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    RMSE = sqrt(mean_absolute_error(y_valid, preds))
    return RMSE
```

## Data Preprocessing

#### Missing Values

```python
#Find Category features and numeric features for data preprocessing
num_cols = X.select_dtypes(exclude=['object']).columns
cat_cols = X.select_dtypes(include=['object']).columns
```

```python
#fill missing values in 'total bedrooms' with Mean using SimpleImputer
from sklearn.impute import SimpleImputer
impute = SimpleImputer(strategy='mean')
impute_num_X_train = pd.DataFrame(impute.fit_transform(X_train[num_cols]))
impute_num_X_valid = pd.DataFrame(impute.transform(X_valid[num_cols]))

#put column names back
impute_num_X_train.columns = X_train[num_cols].columns
impute_num_X_valid.columns = X_valid[num_cols].columns
```

```python
print('imputed num_X_train contains missing values: ' + str(impute_num_X_train.isnull().sum().sum()))
print('imputed num_X_valid contains missing values: ' + str(impute_num_X_valid.isnull().sum().sum()))
```

#### Encode the categorical variables

```python
X_train[cat_cols].head()
```

```python
X_train[cat_cols].value_counts()
```

```python
#Use OneHotEncoder to covert the categorical attribute
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cat_X_train = pd.DataFrame(one_hot_encoder.fit_transform(X_train[cat_cols]))
OH_cat_X_valid = pd.DataFrame(one_hot_encoder.transform(X_valid[cat_cols]))
                              
#one hot encoder removes the index, put it back
#OH_cat_X_train.index = X_train[cat_cols].index
#OH_cat_X_valid.index = X_valid[cat_cols].index    
```

```python
processed_X_train = pd.concat([impute_num_X_train, OH_cat_X_train], axis=1)
processed_X_valid = pd.concat([impute_num_X_valid, OH_cat_X_valid], axis=1)
```

### Construct a Pipeline for Data Preprocessing

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

#Preprocessing for numerical data
num_transformer = SimpleImputer(strategy = 'mean')

#Preprocessing for categorical data
cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown = 'ignore', sparse=False))
])

#Bundle the Preprocessing steps for both numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num_cols', num_transformer, num_cols),
        ('cat_cols', cat_transformer, cat_cols)
    ])
```

### Cross Validation  

```python
from sklearn.model_selection import cross_val_score
#cv_scores = -1 * cross_val_score(final_pipeline, X_train, y_train, cv=5, scoring = 'neg_mean_absolute_error')

#print('Random Forest Model Cross Validation Score :\n ' + str(cv_scores))
#print('Average Cross Validation Score: ' + str(cv_scores.mean()))
```

### Define a get_score function for different model testing 

```python
def get_score (n_estimators):
    final_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('final_model', RandomForestRegressor(n_estimators, n_jobs=32, random_state=0))    
])
    cv_scores = -1 * cross_val_score(final_pipeline, X_train, y_train, cv=5, scoring = 'neg_mean_absolute_error')
    return cv_scores.mean()
```
